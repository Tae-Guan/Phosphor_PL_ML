```python
df = pd.read_csv('/content/original_dataset_3.csv', encoding="UTF-8")
print(df.head())
print(df)
```

```python
x_train = df.drop(['PL wavelength'], axis = 1)
x_train
```

```python
y_train = df.drop(['Crystallite size',	'd(Å)', 'FWHM'], axis = 1)
y_train
```

```python
# imbalanced-learn 패키지
from imblearn.over_sampling import SMOTE

# SMOTE 객체 생성
#smote = SMOTE(random_state=42,k_neighbors=5)
#SMOTE(ratio=0.5,random_state=10)
smote = SMOTE(random_state=42, k_neighbors=1)

# 샘플링
x_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train)
```
